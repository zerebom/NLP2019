```python
import gensim
import pandas as pd 
import os
```

```python
df=pd.read_csv('./形容詞.csv', engine='python')
```

```python
list(df.columns)
```

```python
worddict=df['単語'].to_dict()
```

```python
corpus=[]
for i in range(1,7):
    corpus.append([(a,b) for a,b in df.iloc[:,i].iteritems() if b>=3 and b<=50])

print(corpus)
```

```python
lda=gensim.models.ldamodel.LdaModel(corpus=corpus, num_topics=20, id2word=worddict)
for topic in lda.show_topics(-1):
    print (topic)
for topics_per_document in lda[corpus]:
    print (topics_per_document)
```
