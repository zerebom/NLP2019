```python
#まずはループしないで試してみる。
word=Mdf.iloc[:,0]
a=Mdf.iloc[:,1:7].sum()
Mdf=(Mdf.iloc[:,1:7]*27000/a.values)
Mdf['単語']=word
Mdf=Mdf.ix[:,['単語','先行増','先行減','一致増','一致減','遅行増','遅行減']]

Mdf=Mdf.round(1)


Mdf['先行増/減']=Mdf['先行増']/Mdf['先行減']
Mdf['先行減/増']=Mdf['先行減']/Mdf['先行増']

Mdf['一致増/減']=Mdf['一致増']/Mdf['一致減']
Mdf['一致減/増']=Mdf['一致減']/Mdf['一致増']

Mdf['遅行増/減']=Mdf['遅行増']/Mdf['遅行減']
Mdf['遅行減/増']=Mdf['遅行減']/Mdf['遅行増']

Mdf=Mdf.round(2)

Mdf=Mdf.replace(np.inf,np.nan).fillna(0)
print(Mdf)
```

```python
df3=pd.DataFrame()
for column in['先行増/減','先行減/増','一致増/減','一致減/増','遅行増/減','遅行減/増']:
    df2=Mdf[['単語',column]].sort_values(column,ascending=False).head(20).reset_index()    
    df3=pd.concat([df3,df2],axis=1)
df3=df3.drop(columns='index')
```

```python
df3.to_csv('./末尾尤度比.csv',encoding='shift-jis')
```
